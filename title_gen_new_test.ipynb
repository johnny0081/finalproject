```
gpt2.download_gpt2(model_name="117M")
```

```
file_name = "/content/validation2.txt"
```

```
sess = gpt2.start_tf_sess()
```

```

gpt2.finetune(sess,
              dataset=file_name,
              model_name='117M',
              steps=10000,
              restore_from='fresh',
              run_name='run1',
              print_every=10,
              sample_every=100,
              save_every=100
              )
```

```
while True:
  text = input("text : ")

  inp = '[text]'+text+'\n'+'[summary]\n'

  x = gpt2.generate(sess,
                length=200,
                temperature = 0.6,
                include_prefix=False,
                prefix=inp,
                nsamples=3,
                )
```
